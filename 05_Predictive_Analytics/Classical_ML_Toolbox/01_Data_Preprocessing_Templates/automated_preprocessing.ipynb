# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
x = dataset.iloc[:, :-1].to_numpy()
y = dataset.iloc[:, -1].to_numpy()
x
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

x_train
```

## Taking care of missing data and encoding

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler


numerical_transformer = Pipeline(
    [("impute", SimpleImputer()), ("scale", StandardScaler())]
)
categorical_transformer = Pipeline(
    [("impute", SimpleImputer(strategy="most_frequent")), ("encode", OneHotEncoder())]
)


ct = ColumnTransformer(
    [("cat", categorical_transformer, [0]), ("num", numerical_transformer, [1, 2])],
    remainder="passthrough",
)
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)

np.set_printoptions(suppress=True, precision=4)
x_train
```

Encoding the labels

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
y_train
```
