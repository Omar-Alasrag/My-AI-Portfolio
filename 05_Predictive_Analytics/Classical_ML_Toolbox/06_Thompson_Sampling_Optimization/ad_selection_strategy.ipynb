# Thompson Sampling

```python
# Dataset link
# https://www.kaggle.com/datasets/akram24/ads-ctr-optimisation
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import random
n_ads = dataset.shape[1]

ad_num_selection = []
ads_reward_0 = [0] * n_ads
ads_reward_1 = [0] * n_ads
for i in range(dataset.shape[0]):
    max_num = 0
    selected_ad = 0
    for j in range(n_ads):
        random_num = random.betavariate(ads_reward_1[j]+1, ads_reward_0[j]+1)
        if max_num < random_num:
            max_num = random_num
            selected_ad = j
    reward = dataset.iloc[i, selected_ad]
    if reward == 0:
        ads_reward_0[selected_ad] += 1
    else:
        ads_reward_1[selected_ad] += 1
    ad_num_selection.append(selected_ad)

plt.hist(ad_num_selection)  
            

```
